We first load the data we will use as a Numpy array.

```python
data = np.loadtxt("takens_1.txt")
print(data.shape)
```

Then we make a scatter plot of these data points to see what they actually look like in 2D space.

```python
plot_curve(data, "The raw data points in 2D space", x_label="X", y_label="Y")
```

We can see that they are periodic. Next we will implement time-delay embedding.

As required, we plot the first coordinate against the time in the dataset.

```python
first_coord_against_time = np.vstack((np.arange(data.shape[0]), data[:, 0])).T
plot_curve(first_coord_against_time, title='First coordinate against time', x_label='Time', y_label='First coordinate')
```

```python
takens = TimeDelayEmbedding(data)
```

```python
delay_coords_5 = takens.timedelay(col=0, n=5, target_dim=2)
takens.plot_delay(delay_coords_5, "First coordinate against delayed version with \u0394n = 5.", x_label="x(n)", y_label="x(n+5)")
```

```python
delay_coords_5_twice = takens.timedelay(col=0, n=5, target_dim=3)
takens.plot_delay(delay_coords_5_twice, "First coordinate against two delayed versions with \u0394n = 5.", x_label="x(n)", y_label="x(n+5)", z_label="x(n+10)")
```
