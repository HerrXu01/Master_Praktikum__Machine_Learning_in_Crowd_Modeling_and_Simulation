# Task 3, Training a Variational Auto-Encoder on MNIST

## 1. Initial Setup

For task 3 we used , Numpy and Matplotlib.

```python
import torch; torch.manual_seed(0)
import torch.nn as nn
import torch.nn.functional as F
import torch.utils
import torch.distributions
import torchvision
import numpy as np
import matplotlib.pyplot as plt; plt.rcParams['figure.dpi'] = 200

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

## 2.Define the VAE model

```python
class VariationalEncoder(nn.Module):
    def __init__(self, latent_dims):
        super(VariationalEncoder, self).__init__()
        self.linear1 = nn.Linear(784, 512)
        self.linear2 = nn.Linear(512, latent_dims)
        self.linear3 = nn.Linear(512, latent_dims)

        self.N = torch.distributions.Normal(0, 1)
        self.N.loc = self.N.loc.cpu() # hack to get sampling on the GPU
        self.N.scale = self.N.scale.cpu()
        self.kl = 0

    def forward(self, x):
        x = torch.flatten(x, start_dim=1)
        x = F.relu(self.linear1(x))
        mu =  self.linear2(x)
        sigma = torch.exp(self.linear3(x))
        z = mu + sigma*self.N.sample(mu.shape)
        self.kl = (sigma**2 + mu**2 - torch.log(sigma) - 1/2).sum()
        return z

class Decoder(nn.Module):
    def __init__(self, latent_dims):
        super(Decoder, self).__init__()
        self.linear1 = nn.Linear(latent_dims, 512)
        self.linear2 = nn.Linear(512, 784)

    def forward(self, z):
        z = F.relu(self.linear1(z))
        z = torch.sigmoid(self.linear2(z))
        return z.reshape((-1, 1, 28, 28))

class VariationalAutoencoder(nn.Module):
    def __init__(self, latent_dims):
        super(VariationalAutoencoder, self).__init__()
        self.encoder = VariationalEncoder(latent_dims)
        self.decoder = Decoder(latent_dims)

    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z)

    def generate_samples(self, num_samples):
        z = torch.randn(num_samples,latent_dims).to(device)
        samples = self.decoder(z)
        return samples
```

## 3. Define functions

```python
def train(autoencoder, data, epochs):
    opt = torch.optim.Adam(autoencoder.parameters(),lr=0.001)
    # total_loss = 0
    for epoch in range(epochs):
        for x, y in data:
            x = x.to(device) # CPU
            opt.zero_grad()
            x_hat = autoencoder(x)
            loss = ((x - x_hat)**2).sum()+ autoencoder.encoder.kl
            loss.backward()
            opt.step()
        return autoencoder

def test(autoencoder, data, epochs):
    # model.eval()
    # total_loss = 0
    with torch.no_grad():
        for epoch in range(epochs):
            for x, y in data:
                x = x.to(device) # CPU
                x_hat = autoencoder(x)
                loss = ((x - x_hat)**2).sum()+ autoencoder.encoder.kl
    return autoencoder
```

```python
#plot the latent representation
def plot_latent(autoencoder, data, num_batches=128):
    for i, (x, y) in enumerate(data): 
        z = x.view(-1,784).to(device)
        z = autoencoder.encoder(z) 
        z = z.to('cpu').detach().numpy() 
        plt.scatter(z[:, 0], z[:, 1], c=y, cmap='tab10') 
    plt.colorbar()
    plt.title('Latent Representation')
    plt.xlabel('Latent Dimension 1')
    plt.ylabel('Latent Dimension 2')
```

```python
#plot reconstructed digits
def plot_reconstructed(autoencoder, data, num_digits=15):
    for batch in data:
        x, y = batch
        break 
    x = x.to(device) 
    z = autoencoder(x)
    x = x.view(-1, 28, 28).cpu().numpy() 
    z = z.view(-1, 28, 28).cpu().detach().numpy()  
    
    fig, axes = plt.subplots(3, 5,figsize=(8, 4))
    for i in range(15):
        row = i // 5  # Calculate the row index
        col = i % 5   # Calculate the column index
        axes[row, col].imshow(x[i])  # Display the image
        axes[row, col].axis('off')   
    plt.suptitle('Raw Digits')
        
    fig, axes = plt.subplots(3, 5,figsize=(8, 4))
    for i in range(15):
        row = i // 5  # Calculate the row index
        col = i % 5   # Calculate the column index
        axes[row, col].imshow(z[i])  # Display the image
        axes[row, col].axis('off') 
    plt.suptitle('Reconstructed Digits')
    plt.show()
```

```python
#plot reconstructed digits
def plot_generated(autoencoder, num_samples=15):
    z = autoencoder.generate_samples(num_samples) 
    z = z.view(-1, 28, 28).cpu().detach().numpy() 

    fig, axes = plt.subplots(3, 5,figsize=(8, 4))
    for i in range(15):
        row = i // 5  # Calculate the row index
        col = i % 5   # Calculate the column index
        axes[row, col].imshow(z[i])  # Display the image
        axes[row, col].axis('off') 

    plt.suptitle('Generated Digits')
    plt.show()
```

## 4.Dataset

```python
train_data = torch.utils.data.DataLoader(
        torchvision.datasets.MNIST('./data',
                                   train=True,
               transform=torchvision.transforms.ToTensor(),
               download=True),
        batch_size=128,
        shuffle=True)

test_data = torch.utils.data.DataLoader(
        torchvision.datasets.MNIST('./data',
                                   train=False,
               transform=torchvision.transforms.ToTensor(),
               download=True),
        batch_size=128,
        shuffle=False)
```

```python
latent_dims = 2
vae = VariationalAutoencoder(latent_dims).to(device) # CPU
```

```python
vae = train(vae, train_data,1)
plot_latent(vae, test_data)
plot_reconstructed(vae, test_data)
plot_generated(vae)
```

```python
# Function to plot the loss curve
def plot_loss_curve(train_losses, test_losses):
    plt.figure(figsize=(8, 6))
    epochs = len(train_losses)
    plt.plot(range(1, epochs+1), train_losses, label='Train')
    # plt.plot(range(1, epochs+1), test_losses, label='Test')
    plt.xlabel('Epoch')
    plt.ylabel('ELBO Loss')
    plt.title('Loss Curve')
    plt.legend()
    plt.show()
```

```python
# Plot the loss curve
plot_loss_curve(train_losses, test_losses)
```
