```python
import json
import matplotlib.pyplot as plt
```

```python
def load_data_from_json(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data
```

```python
nn_file_path = 'nn_results.json'
fd_file_path = 'fd_results.json'
```

```python
nn_results = load_data_from_json(nn_file_path)
fd_results = load_data_from_json(fd_file_path)
```

```python
fig, ax = plt.subplots()

ax.plot(fd_results.keys(), fd_results.values(), label='FD', color='red', marker='o')
ax.plot(nn_results.keys(), nn_results.values(), label='ANN', color='gray', marker='o')

ax.legend()

ax.set_title('Test error of NN and FD', fontsize=20)
ax.set_xlabel('Experiments', fontsize=18)
ax.set_ylabel('MSE', fontsize=18)

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
