## Define Parameters for Phase Portrait

```python
#This lines set the parameter of the phase portrait
w = 5
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
```

## Main Function

```python
def draw_phase_potrait(A,alpha,beta):
    #Euler's method to construct and plot a trajectory over the stream plot
    y0 = np.array([3, 3])
    time = np.linspace(0, 10, 100)
    yt, time = solve_euler(lambda y: A@y, y0, time)

    #Linear vector field A*x
    ax0 = plot_phase_portrait(A, X, Y)

    #Gets the eigenvalues of the matrix A
    eigvals = np.linalg.eigvals(A)
    
    #set the title of our phase portrait
    ax0.set_title(f"alpha: {alpha}, beta: {beta}, Lambda 1: {eigvals[0]:.4f}, Lambda 2: {eigvals[1]:.4f}")

    #then plot the trajectory over it
    #ax0.plot(yt[:, 0], yt[:, 1], c='red', label='Example trajectory')

    # prettify
    ax0.legend()
    ax0.set_aspect(1)
```

## Matrices, Parameters and Phase Portraits

### Stable Node

```python
alpha = -1
beta = -3
A = np.array([
    [alpha, 0],
    [0, beta]
])
draw_phase_potrait(A,alpha,beta)
```

### Stable Focus

```python
alpha = -0.1
beta = 0.1
A = np.array([
    [alpha, beta],
    [-1/4, 0]
])
draw_phase_potrait(A,alpha,beta)
```

### Unstable Saddle

```python
alpha = 1
beta = -1
A = np.array([
    [alpha, 0],
    [0, beta]
])
draw_phase_potrait(A,alpha,beta)
```

### Unstable Node
