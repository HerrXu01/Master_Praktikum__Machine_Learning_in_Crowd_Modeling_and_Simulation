```python
from lorenz import *
```

We define a function `lorenz_simulate`, in which we call `odient` function to numerically solve the lorenz system given an initial state and the time points.  

We also plot the trajectories in 3D space.  

We first plot the trajectory of the lorenz attractor with initial point (10, 10, 10) and T_end = 1000.

```python
lorenz_simulate(x0=[10.0, 10.0, 10.0])
```

Then in order to see how a small perturbation will affect the trajectory, we plot the trajectory of the lorenz attractor with initial point (10+1e-8, 10, 10) and T_end = 1000.

```python
lorenz_simulate(x0=[10.0 + 1e-8, 10.0, 10.0])
```

We define a function `lorenz_compare` to plot the two trajectories together for a more intuitive comparison.  

Additionally, we plot the L2 norm of the difference of two trajectories against time. We add a red dash line representing 1 to see at which time points the difference are larger than 1. We also print the first time point where the difference is larger than 1.

```python
lorenz_compare(x0=[10.0, 10.0, 10.0], x0_hat=[10.0 + 1e-8, 10.0, 10.0])
```

Now we change the parameter `rho` to the 0.5 and plot the two trajectories.

```python
lorenz_simulate(x0=[10.0, 10.0, 10.0], rho=0.5)
```

```python
lorenz_simulate(x0=[10.0 + 1e-8, 10.0, 10.0], rho=0.5)
```

```python
lorenz_compare(x0=[10.0, 10.0, 10.0], x0_hat=[10.0 + 1e-8, 10.0, 10.0], rho=0.5)
```
