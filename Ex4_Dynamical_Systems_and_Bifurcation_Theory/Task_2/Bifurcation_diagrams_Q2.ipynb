```python
import numpy as np
import matplotlib.pyplot as plt

def plot_curves(beta, gamma):
    """
        Args: 
            beta(int): fixed parameter of the system, assumed as a poitive quantity
            gamma(int): fixed parameter of the system
            
        Returns:
            Plots the bifurcation diagram for a given dynamical system
            
    """
    # generating alpha values
    alpha = np.linspace(-gamma , 8 - gamma, 500)  #a range big enough

    # calculating the respective state values (x) for each alpha
    y_positive = np.sqrt((alpha + gamma)/beta)
    y_negative = -np.sqrt((alpha + gamma)/beta)

    # plotting curves
    plt.figure(figsize=(4, 4))
    plt.plot(alpha, y_positive, 'g', label = "stable equilibrium")
    plt.plot(alpha, y_negative, 'r', label = "unstable equilibrium")
    plt.plot(-gamma,0, marker = 'o', color = 'black', label = 'saddle point')

    # adding a range to the values of alpha
    plt.xlim(-gamma - 3, -gamma + 5)

    # adding legend
    plt.legend()

    plt.grid(True)
    plt.show()
```

Here the curves are generalised as dx/dt = alpha - beta * x^2 + gamma
*alpha is still the only parameter for a system which will vary, beta and gamma remain constant for 1 system throughout.

```python
# calling the function to plot the curves

plot_curves(1,0) # System - 1
plot_curves(2,-3) # System - 2
```
